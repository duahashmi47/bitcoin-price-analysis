```python
from pycoingecko import CoinGeckoAPI
```

```python
cg = CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency = 'usd', days =30)
```

```python
bitcoin_data['prices']
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(bitcoin_data['prices'], columns=['TimeStamp','Price'])
```

```python
data.head()
```

```python
data.shape
```

```python
data['Date'] = pd.to_datetime(data['TimeStamp'], unit = 'ms')
```

```python
data.shape
```

```python
data.head()
```

```python
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price':['min','max','first','last']})
```

```python
candlestick_data
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=[go.Candlestick(x =candlestick_data.index, open = candlestick_data['Price']['first'],
                                                      high = candlestick_data['Price']['max'], low = candlestick_data['Price']['min'],
                                                      close = candlestick_data['Price']['last'])])
```

```python
fig.update_layout(xaxis_rangeslider_visible = False, xaxis_title = 'Date',
                 yaxis_title = 'Price (USD $)', title = 'Bitcoin Candlestick Chart Over Past 30 Days')
```

```python
import plotly
```

```python
plotly.offline.plot(fig, filename = './bitcoin_candlestick_graph.html', auto_open = False)
```
